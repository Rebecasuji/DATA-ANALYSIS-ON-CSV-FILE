```python
import pandas as pd

data = {
    'Date': ['2025-01-01', '2025-01-02', '2025-01-03', '2025-01-04', '2025-01-05', '2025-01-06'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South'],
    'Product': ['Product A', 'Product B', 'Product A', 'Product C', 'Product B', 'Product A'],
    'Sales': [200, 150, 180, 100, 300, 250]
}

df = pd.DataFrame(data)
df.to_csv('sales_data.csv', index=False)
print("CSV file created")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv('sales_data.csv')

# Preview
df.head()
```

```python
# Grouping and summarizing
sales_by_region = df.groupby('Region')['Sales'].sum()
sales_by_product = df.groupby('Product')['Sales'].sum()

# Plotting
sales_by_region.plot(kind='bar', title='Sales by Region', color='skyblue')
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.grid()
plt.show()

sales_by_product.plot(kind='pie', title='Sales by Product', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```
